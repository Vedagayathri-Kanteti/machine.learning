```python
import numpy as np

arr = np.arange(1,21)
print(arr)
print("The shape of array is:", arr.shape)
print("The minimum value in array is:", min(arr))
print("The Maximum value in atrray is:" ,max(arr))
mean = np.mean(arr)
print("The mean of the array is:" , mean)
stand = np.std(arr)
print("The standard deviation of the array is:", stand)
arr_1 = arr.reshape(4,5)
print("the new shape of array is:\n", arr_1)
print("the second row of new shaped array is:")
for i in range(0,5):
    print(arr_1[1][i], end = " ")
print("\nthe third column of new shaped array is:")
for j in range(0,4):
    print(arr_1[j][2])
```

```python
import numpy as np
A = np.array([10,20,30,40,50])
B = np.array([5,10,15,20,25])
addn = [x + y for x,y in zip(A,B)]
print("The sum of both sets:", addn)
subt = [y - x for y,x in zip(B,A)]
print("Subtraction from b to a:" , subt)
mult = [x*y for x,y in zip(A,B)]
print("Multiplication:", mult)
div = [x/y for x,y in zip(A,B)]
print("Division:", div)
sum = 0
for i in range(0,4):
    sum += A[i]
print("The total sum of A:" , sum)
```

```python
import numpy as np
import pandas as pd

data = {"Roll No" : [101,102,103,104,105],
        "Name" : ["Ravi", "Priya" , "Kiran", "Anu", "Sita"],
        "Marks" : [85,92,78,88,95]}
df = pd.DataFrame(data)
display(df)
display("First three records:", df[0:3])
print("the average marks:", df['Marks'].mean())
print("The person with highest marks:", df.loc[df['Marks'].idxmax(), 'Name'])
print("The person with the lowest marks:", df.loc[df['Marks'].idxmin(), 'Name'])
df['Grade'] = ['B','A','C','B','A']
display(df)
```

```python
import numpy as np
import pandas as pd

data = {'Product' : ['Laptop','Mouse','Keyboard','Monitor'],
        'Price' : [50000,500,1200,10000],
        'Quantity' : [10,50,30,15]}
df = pd.DataFrame(data)
display(df)
df['Total Value'] = df['Price'] * df['Quantity']
display(df)
high = df.loc[df['Total Value'].idxmax(),'Product']
print("The highest total value product is:",high)

total = (df['Total Value']).sum()
print("The total inventory value is:",total)
```
